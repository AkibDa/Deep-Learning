# Classifying Handwritten Digits with a Neural Network

```python
import tensorflow as tf
```

```python
mnist = tf.keras.datasets.mnist

(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train.shape
```

```python
y_train.shape
```

```python
X_train, X_test = X_train / 255.0, X_test / 255.0

X_train[0]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Input, Dropout

model = Sequential([
    Input(shape=(28, 28)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(10)
])
```

```python
nonsense_predictions = model(X_train[:1]).numpy()
nonsense_predictions
```

```python
tf.nn.softmax(nonsense_predictions).numpy().sum()
```

```python
from tensorflow.keras.losses import SparseCategoricalCrossentropy

loss_fn = SparseCategoricalCrossentropy(from_logits=True)
loss_fn(y_train[:1], nonsense_predictions).numpy()
```

```python
from tensorflow.keras.optimizers import Adam

optimizer = Adam(learning_rate=0.01)

model.compile(optimizer=optimizer,
              loss=loss_fn,
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=10)
```

```python
model.evaluate(X_test, y_test)
```
